```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

```

```python

x = np.arange(0,1,100)

y = np.sin(x)
z = np.cos(x)

```

```python
f, axarr = plt.subplots(1,2)
axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

#rightside
axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

```
